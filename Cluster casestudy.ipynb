# KMeans-Clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',n_init=20)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel('no.of clusters')
plt.ylabel('wcss')
plt.show()
```
